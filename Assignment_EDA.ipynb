### FEATURES
The various features of the cleaned dataset are explained below:
1) show_id: The unique id of the netflix show is stored in the show_id column. It is a categorical feature.
2) type: type stores information regarding the netflix show type, whether it's a movie or TV show. It is a categorical feature.
3) title: title stores the name of the show. It is a categorical feature.
4) director: director stores the director of the show. It is a categorical feature.
5) cast: cast stores the main actors in the show. It is a categorical feature. 
6) country: country stores where the show was produced. It is a categorical feature.
7) date_added: date when the show was added to Netflix. It is a categorical feature.
8) release_year: year the show was released. It is a numerical feature.
9) rating: A categorical feature that classifies the sjows based on the age of the viewers.
10) duration: A continuous feature that displays length of the movie or number of seasons for a TV show.
11) listed_in: listed_in stores the genres or the categories the show belongs to. It is a categorical feature.
12) description: Target variable that stores short summary of the show’s content.

**COLUMN BREAKDOWN:**

show_id – Unique ID.

type – Movie or TV show.

title – Name of the content.

director – Director(s).

cast – Main actors.

country – Production country.

date_added – Netflix release date.

release_year – Original release year (Numeric).

rating – Age rating.

duration – Length (e.g., "90 min", "2 Seasons").

listed_in – Genre(s).

description – Brief summary.

## Handling Categorical values by taking the mode

```python
df['director'].fillna(df['director'].mode()[0], inplace=True)
df['cast'].fillna(df['cast'].mode()[0], inplace=True)
df['country'].fillna(df['country'].mode()[0], inplace=True)
df['date_added'].fillna(df['date_added'].mode()[0], inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
df['duration'].fillna(df['duration'].mode()[0], inplace=True)
```

## Feature Engineering

```python
df['Day_Added'] = df['date_added'].str.split(' ').str[1].str.replace(',', '')
df['Month_Added'] = df['date_added'].str.split(' ').str[0]
df['Year_Added'] = df['date_added'].str.split(' ').str[2]
```

```python
df['Day_Added'] = df['Day_Added'].astype(int)
df['Year_Added'] = df['Year_Added'].astype(int)
```

## Multivariate using pie plots

```python
plt.rcParams['figure.figsize'] = (20, 10)

plt.subplot(1, 5, 1)
size = df['Month_Added'].value_counts()
labels = size.index  
color = ['salmon', 'teal', 'orange'][:len(labels)]  

plt.pie(size, colors=color, labels=labels, autopct='%1.2f%%')
plt.title('Month_Added', fontsize=20)
plt.axis('off')


plt.subplot(1, 5, 2)
size = df['type'].value_counts()
labels = size.index  
color = ['pink','lightgreen']

plt.pie(size, colors=color, labels=labels, autopct='%1.2f%%')
plt.title('type', fontsize=20)
plt.axis('off')



plt.tight_layout()
plt.show()
```

```python
f, ax = plt.subplots(1, 2, figsize=(20, 10))

sns.countplot(x=df['Month_Added'], hue=df['Month_Added'], data=df, palette='pastel', ax=ax[0], saturation=0.95, legend=False)

ax[0].set_title('Count of Month_Added')

for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=20)

size = df['Month_Added'].value_counts()
labels = size.index  
colors = ['green', 'hotpink', 'lightblue'][:len(labels)]  
explode = [0.1 if i == 1 else 0 for i in range(len(labels))]  

ax[1].pie(size, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, colors=colors)
ax[1].set_title('Distribution of Month_Added')

plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python

# Initialize the encoder
encoder = OneHotEncoder(sparse_output=False, drop='first')  # drop='first' prevents redundancy

# Select categorical columns to encode
categorical_cols = ['type', 'country', 'listed_in']

# Encode the categorical columns
encoded_cols = pd.DataFrame(
    encoder.fit_transform(df[categorical_cols]),
    columns=encoder.get_feature_names_out(categorical_cols)  # Get meaningful column names
)

# Reset the index to match df
encoded_cols.index = df.index  

# Drop original categorical columns
df.drop(columns=categorical_cols, inplace=True)

# Concatenate the encoded features with the original dataframe
df = pd.concat([df, encoded_cols], axis=1)

# Display the updated dataset
print(df.head())
```
