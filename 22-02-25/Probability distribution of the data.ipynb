```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

p=0.6

data=bernoulli.rvs(p, size=1000)

plt.bar([0,1],[sum(data==0), sum(data==1)], color=['hotpink','lightgreen'], tick_label=["Failure (0)", "Success (1))"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Berboulli Distribution (p=0.6)")
plt.show()
```

# Implementing Poisson Distribuion in python

```python
from scipy.stats import poisson

l = 3

x = np.arange(0, 10)
pmf = poisson.pmf(x , l)

plt.bar(x, pmf, color='orange', alpha=0.7)
plt.xlabel("Number of calls per Minute")
plt.ylabel("Probability")
plt.title(f"Poisson Distribution (l={l})")
plt.show()
```

# Normal Gaussian Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 70
sigma = 10

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, color='purple', label=f"N({mu}, {sigma}^2)")
plt.xlabel("Exam scores")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (={mu}, sigma={sigma})")
plt.legend()
plt.grid()
plt.show()
```

# Implementing Standard normal distribution in python

```python
from scipy.stats import norm

z = np.linspace(-4, 4, 1000)
pdf = norm.pdf(z, 0, 1)

plt.plot(z, pdf, color='red', label="Standard Normal Distribution (=0, sigma=1)")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.title("Standard Normal Distribution (Z-Distribution)")
plt.axvline(x=0, color='green', linestyle='--', label="Mean {z=0}")
plt.legend()
plt.grid()
plt.show()
```

```python
import seaborn as sns 

np.random.seed(42)

income_data = np.random.exponential(scale=50, size=1000)

plt.figure(figsize=(8, 4))
sns.histplot(income_data, bins=50, kde=True, color='violet')
plt.title("Original Income Distribution (skewed)")
plt.xlabel("Income ($1000)")
plt.ylabel("Frequency")
plt.show()

sample_means=[]
n_samples = 1000
sample_size = 30

for _ in range(n_samples):
    sample = np.random.choice(income_data, size=sample_size)
    sample_means.append(np.mean(sample))
    
plt.figure(figsize=(8,4))
sns.histplot(sample_means, bins=50, kde=True, color='b')
plt.title(f"Sampling Distribution of Sample Mean (n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
