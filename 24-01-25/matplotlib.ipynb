```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/CVR/Downloads/clevelanda (1).csv')
```

```python
df
```

```python
month = ['jan', 'feb', 'march', 'april', 'may' , 'june' , 'july' , 'august' , 'sept' , 'oct' , 'nov' , 'dec']
salery = [1000, 1200, 1100, 1150, 1300, 1400, 1320, 1234, 2345, 1111, 2345, 1265]
plt.plot(month[:6],salery[:6],color='green',linestyle='dashed',marker='*')
plt.title('Line plot')
plt.xlabel('month')
plt.ylabel('salery')
plt.show()
```

```python
plt.bar(df['age'],df['oldpeak'],color='hotpink')
plt.title('Line plot')
plt.xlabel('age')
plt.ylabel('oldpeak')
plt.show()
```

```python
plt.scatter(df['trestbps'],df['chol'],color='purple')
plt.title('Line plot')
plt.xlabel('trestbps')
plt.ylabel('chol')
plt.show()
```

```python
categories = ['A','B','C','D','E']
values = [10,15,7,12,9]
plt.bar(categories, values, color=['olive', 'beige', 'maroon', 'coral', 'gold'])
plt.title('Line plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
values = [12,15,7,12,9,23,43,44,23,43]
plt.hist(values, 6, color='teal',edgecolor='black')
plt.title('Histogram')
plt.show()
```

```python
sizes = [30, 25, 20, 25]
labels = ['Apple', 'Banana', 'Orange', 'Grapes']
plt.pie(sizes, labels=labels, colors=['#c2c2f0', '#ffb3e6', '#c4e17f', '#f7a700'], autopct='%1.1f%%')
plt.show()
```
