```python
from sklearn.preprocessing import LabelEncoder

data = {'Difficulty' : ['Easy', 'Medium', 'Hard', 'Medium', 'Easy'],
        'Test Score' : [50,60,70,80,90]}
df = pd.DataFrame(data)
encoder = LabelEncoder()
df['Difficulty_encoded'] = encoder.fit_transform(df['Difficulty'])
print(df, '\n')
corr = df['Difficulty_encoded'].corr(df['Test Score'])
print('Correlation : ', corr)
```

```python
data = {'Color' : ['Red', 'Blue', 'Green', 'Blue', 'Red'],
        'Test Score' : [50,60,70,80,90]}
df = pd.DataFrame(data)
df_encoded = pd.get_dummies(df, columns = ['Color'])
print(df_encoded, '\n')
corr = df_encoded['Color_Red'].corr(df_encoded['Test Score'])
print('Correlation with Red : ', corr)
```
