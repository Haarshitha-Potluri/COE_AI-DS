```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/CVR/Downloads/titanic/train.csv')
```

```python
df
```

```python
sns.barplot(data=df, x='Pclass',y='Survived',palette=['tomato','coral','salmon'],errorbar=None)
plt.title("Survival Rate by Passenger class")
plt.xlabel('Passenger')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Fare',hue='Sex',palette=['green','hotPink'], alpha=0.35)
plt.title('Scatter Plot of Age vs. Fare')
plt.show()
```

```python
sns.pairplot(data=df,vars=['Age', 'Fare', 'Pclass'],hue='Pclass',palette='Greens')
plt.show()
```

```python
numeric_train = df.select_dtypes(include=['number'])
c = numeric_train.corr()
sns.heatmap(c,annot=True,cmap='Greens',fmt='.1f')
plt.title('Correlation Heat map')
plt.show()
```
