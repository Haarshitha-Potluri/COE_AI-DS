```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

```python
sns.histplot(data=tips,x='total_bill',kde=True,bins=20,color='gold')
plt.title('Distribution of total bills')
plt.show()
```

```python
sns.kdeplot(data=tips,x='total_bill',fill=True,color='salmon')
plt.title('KDE plot of total bills')
plt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill',palette='Pastel1')
plt.title('Average Total bill by day')
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',palette='PiYG')
plt.title('Box Plot of total bills by day and gender')
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',palette='Paired')
plt.title('Scatter Plot of total bill vs. tip')
plt.show()
```

```python
sns.lineplot(data=tips,x='size',y='tip',marker='o',palette='Set1')
plt.title('Line Plot of Party size vs. tip')
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='Set2')
plt.show()
```

```python
numeric_tips = tips.select_dtypes(include=['number'])
c = numeric_tips.corr()
sns.heatmap(c,annot=True,cmap='magma',fmt='.4f')
plt.title('Correlation Heat map')
plt.show()
```
