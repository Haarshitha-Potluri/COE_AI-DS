```python
import numpy as np
```

```python
def LinearRegression(x,y):
    x_m = np.mean(x)
    y_m = np.mean(y)
    num = 0
    den = 0
    
    n=len(x)
    for i in range(n):
        num = num + (x[i] - x_m) * (y[i]-y_m)
        den = den + (x[i] - x_m)**2
    
    m=num/den
    c=y_m -m * x_m
    return m,c
    
    
x=[1,2,3,4,5]
y=[3,4,2,4,5]
```

```python
m,c = LinearRegression(x,y)
print(f"y = {m}x + {c}")
```

```python
def Predict(x):
    y_p = []
    for i in range(len(x)):
        y_p.append(m * x[i] + c)
    return y_p

y_p = Predict(x)
print(y_p)
```

```python
y_m = np.mean(y)
r2 = 1 - ((sum(y_p - y_m) ** 2) / (sum(y - y_m) ** 2))
print(f"r2 = {r2}")
```

```python
y, y_p
```

```python
def r2_score(y,y_p):
    y_m = np.mean(y)
    print(f"Mean of y : {y_m}")
    num = 0
    den = 0
    
    for i in range(len(y)):
        num = num + ((y_p - y_m) ** 2)
        den = den + ((y - y_m) ** 2)
    return num/den
```

```python
print("r2 score : ",r2_score(y,y_p))
```

```python
print("Accuracy : ",1-r2_score(y,y_p))
```
